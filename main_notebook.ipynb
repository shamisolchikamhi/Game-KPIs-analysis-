```python
import pandas as pd
from data_wrangling import DataCleaning, adspend_df, installs_df, payouts_df, revenue_df
from data_analytics import DataAnalytics
from data_analytics import ChartMaker
import matplotlib.pyplot as plt
```

```python
analytics = DataAnalytics(adspend_df=adspend_df, installs_df=installs_df, payouts_df=payouts_df, revenue_df=revenue_df)
```

```python
ChartMaker.hist_chart(data=installs_df, column='network_id', title='Users per network')
ChartMaker.time_series_breakdown(data = installs_df.groupby(['network_id', 'year and month'])['install_id'].count().reset_index(),
                             x_column='year and month', y_column='install_id', breakdown_column= 'network_id', title='Users per network over time')
```

```python
ChartMaker.bar_chart(analytics.user_acquisition_costs(groupby_column = 'network_id'), x_column='network_id',
                     y_column='user_acquisition_cost_usd', title='User aquisation cost per network')
ChartMaker.time_series_breakdown(analytics.user_acquisition_costs(groupby_column = ['network_id','event_date']), x_column='event_date',
                     y_column='user_acquisition_cost_usd', breakdown_column='network_id', title='User aquisation cost per network')
```

```python
ChartMaker.pie_chart(data= analytics.revenue_generated_per_install(groupby_column='network_id'),
                     label_column= 'network_id', value_column= 'revenue_per_install_usd',
                     title='Average Revenue per network')
ChartMaker.time_series_breakdown(analytics.revenue_generated_per_install(groupby_column = ['network_id','event_date_x']),
                                 x_column='event_date_x',
                     y_column='revenue_per_install_usd', breakdown_column='network_id', title='Average Revenue per network over time')
```

```python
ChartMaker.pie_chart(data= analytics.total_payouts_made_per_install(groupby_column='network_id'),
                     label_column= 'network_id', value_column= 'payouts_per_install_usd',
                     title='Average Revenue per network')
ChartMaker.time_series_breakdown(analytics.total_payouts_made_per_install(groupby_column = ['network_id','event_date_x']),
                                 x_column='event_date_x',
                     y_column='payouts_per_install_usd', breakdown_column='network_id', title='Average Revenue per network over time')
```
